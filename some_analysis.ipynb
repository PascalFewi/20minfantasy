```python
import requests
import json
import pandas as pd

url = "https://fantasy.20min.ch/json/fantasy/stats_centre_table.json"

try:
    response = requests.get(url)
    response.raise_for_status()  # Raise an exception for bad status codes
    data = response.json()
    df = pd.DataFrame(data)
    # You can now work with the 'data' variable which contains the JSON content
    # print(data) # Uncomment this line to see the data
except requests.exceptions.RequestException as e:
    print(f"Error fetching data: {e}")
except json.JSONDecodeError as e:
    print(f"Error decoding JSON: {e}")

print(f"Anzahl Spieler: {len(data)}")
print(df.columns)
```

Nun möchten wir das perfekte Team finden. Dafür vereinfachen wir die Constraints etwas:
Da Stürmer im Vergleich zu anderen Positionen deutlich schwächer in Points per Price und Average Points per Price abschneiden, nehmen wir an, dass die optimale Formation entweder ein 5-4-1 oder ein 4-5-1 ist.

Wir versuchen also, die besten 11 Spieler zu finden und die übrigen Positionen mit den günstigsten verfügbaren Spielern aufzufüllen. Im Sturm gibt es nur einen 4-Millionen-Spieler und einige wenige für 4.5 Millionen. Ein schlechter Torwart kostet uns rund 3.5 Millionen. Ergänzt wird die Ersatzbank mit einem beliebigen 4-Millionen-Spieler aus dem Mittelfeld oder der Verteidigung (abhängig von der gewählten Formation).

Daraus ergibt sich:
16 Millionen werden für günstige Ersatzspieler verwendet.
Damit bleiben 84 Millionen für unsere optimale Startelf übrig.

Für diese Optimierung wenden wir den Knapsack-Algorithmus an mit den folgenden Constraints:

    Maximal 5 Verteidiger

    Maximal 5 Mittelfeldspieler

    Insgesamt maximal 9 Spieler aus Verteidigung + Mittelfeld

    Genau 1 Torwart

    Genau 1 Stürmer

    Maximal 3 Spieler pro Team

Ziel: Maximiere die Total Points der Startaufstellung.

```python
import pulp
from collections import Counter

# Problem definieren
problem = pulp.LpProblem("Best_Fantasy_Team", pulp.LpMaximize)

# Binärvariable: 1 = Spieler im Team, 0 = nicht im Team
player_vars = {
    row['displayName']: pulp.LpVariable(row['displayName'], cat='Binary')
    for _, row in df.iterrows()
}

# Zielfunktion: Maximiere totalPoints
problem += pulp.lpSum([
    player_vars[row['displayName']] * row['totalPoints']
    for _, row in df.iterrows()
]), "Total_Team_Points"

# Budget-Beschränkung (max 84 Mio)
problem += pulp.lpSum([
    player_vars[row['displayName']] * row['price']
    for _, row in df.iterrows()
]) <= 84, "Budget"

# Genau 11 Spieler
problem += pulp.lpSum(player_vars.values()) == 11, "Total_Players"

#  Genau 1 Torwart
problem += pulp.lpSum([
    player_vars[row['displayName']]
    for _, row in df[df['position'] == 'GK'].iterrows()
]) == 1, "One_Goalkeeper"

# Genau 1 Stürmer
problem += pulp.lpSum([
    player_vars[row['displayName']]
    for _, row in df[df['position'] == 'FWD'].iterrows()
]) == 1, "One_Forward"

# Max 5 DEF + Max 5 MID + Max 9 DEF+MID
problem += pulp.lpSum([
    player_vars[row['displayName']]
    for _, row in df[df['position'] == 'DEF'].iterrows()
]) <= 5, "Max_5_DEF"

problem += pulp.lpSum([
    player_vars[row['displayName']]
    for _, row in df[df['position'] == 'MID'].iterrows()
]) <= 5, "Max_5_MID"

problem += pulp.lpSum([
    player_vars[row['displayName']]
    for _, row in df[df['position'].isin(['DEF', 'MID'])].iterrows()
]) <= 9, "Max_9_DEF_MID"

# 6️Max 3 Spieler pro Team
for team in df['squadId'].unique():
    problem += pulp.lpSum([
        player_vars[row['displayName']]
        for _, row in df[df['squadId'] == team].iterrows()
    ]) <= 3, f"Max_3_from_{team}"
```

```python
# Problem lösen
solver = pulp.PULP_CBC_CMD(msg=False)
problem.solve(solver)

# Ergebnis anzeigen
chosen_players = []
for _, row in df.iterrows():
    name = row['displayName']
    if pulp.value(player_vars[name]) == 1:
        chosen_players.append(row)

team_df = pd.DataFrame(chosen_players)
total_cost = team_df['price'].sum()
total_points = team_df['totalPoints'].sum()

print(f"Kosten: {total_cost:.2f} Mio")
print(f"Total Points: {total_points:.0f}")
display(team_df[['displayName', 'position', 'squadId', 'price','averagePoints', 'totalPoints']].sort_values(by='position'))
```
